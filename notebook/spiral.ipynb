```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
data = []
prediction = []
with open("../spiral_data") as f:
    for line in f:
        sp = line.split(" ")
        if sp[0] == "data":
            label, x, y, c = sp
            data.append([float(x), float(y), int(c)])
        elif sp[0] == "prediction":
            label, x, y, c = sp
            prediction.append([float(x), float(y), int(c)])

fig, ax = plt.subplots(1)

x = sorted(list(set([x for x,y,c in prediction])))
Y, X = np.meshgrid(x, x)
Z = np.array([c for x,y,c in prediction])
Z = Z.reshape(X.shape)
ax.contourf(X, Y, Z, cmap=plt.cm.Spectral, alpha=0.8)

ax.scatter(
    [x for x,y,c in data], 
    [y for x,y,c in data], 
    c=[c for x,y,c in data],
    cmap=plt.cm.Spectral)

ax.axis('scaled')
```
